```python
def matxbuida ():
    """
    
    Returns
    -------
    mat : defined matrix of 6x7.

    """
    n_files = 6
    n_columnes = 7


    mat = []

    for i in range (n_files):
        mat.append ([])
        
        
    for i in range (n_files):
        
        for j in range (n_columnes):
            mat[i].append(" _ ")
            
    return mat

def taulell (mat):
    """
    
    Description
    ----------
    Gives visual format to the matrix list.
    
    Parameters
    ----------
    mat : matrix (list).

    Returns
    -------
    mat : matrix (list).

    """
    for i in range (0,6):
        print ("")
        for j in range (0,7):
            print (mat[i][j], end="")
    return mat

def jugador1 (mat):
    """
    
    Description
    ----------
    Asks for a column and shows X at the last empty row
    Full column considered (missage)
    Row out of range considered
    
    Parameters
    ----------
    j : column (int)
    
    mat : matrix (list)

    Returns
    -------
    mat : modified matrix.

    """
    
    i = 5
    error = True
    
    
    while error == True:
        
        j = input ("columna:")
        
        if j in "1234567":
            j = int (j) - 1
            
            if j < 7 and mat [0][j] == " _ ":
                error = False
            else:
                print ("Dona'm una columna correcta jefe!!🧐")
        
        else:
            print ("Dona'm una columna correcta jefe!!🧐")
        
        
    while i >= 0:
        if mat [i][j] == " _ ":
            mat[i][j] = " X "
            break
        i = i - 1
    
    return mat
        
def jugador2 (mat):
    """
    
    Description
    ----------
    Shows O at the last empty random row of a random column
    Full column considered
    
    Parameters
    ----------
    j : random column (int)
    
    mat : matrix (list)

    Returns
    -------
    mat : modified matrix.


    """
    
    import random
    i = 5
    j = random.randint (0,6)
    
    while mat [0][j] != " _ ":
        j = random.randint (0,6)
        
    while i >= 0:
        if mat [i][j] == " _ ":
            mat[i][j] = " O "
            break
        i = i - 1
    
    return mat
        
def has_guanyat_files (mat):
    """
    Description
    ----------
    Reads matrix and looks for 4 equivalent row positions
    
    Parameters
    ----------
    mat : matrix (list).

    Returns
    -------
    estat : True (4 equivalent positions)
            False (rest)

    """
    
    estat = False
    
    for i in range (0,6):
        for j in range (0,4):
            if mat [i][j] == mat [i][j+1] and mat [i][j] != " _ ":
                if mat [i][j] == mat [i][j+2] and mat [i][j] != " _ ":
                    if mat [i][j] == mat [i][j+3] and mat [i][j] != " _ ":
                        estat = True
    
    return estat

def has_guanyat_cols (mat):
    """
    
    Description
    ----------
    Reads matrix and looks for 4 equivalent column positions
    
    Parameters
    ----------
    mat : matrix (list).

    Returns
    -------
    estat : True (4 equivalent positions)
            False (rest)

    """
    
    estat = False
    
    for j in range (0,7):
        for i in range (0,3):
            if mat [i][j] == mat [i+1][j] and mat [i][j] != " _ ":
                if mat [i][j] == mat [i+2][j] and mat [i][j] != " _ ":
                    if mat [i][j] == mat [i+3][j] and mat [i][j] != " _ ":
                        estat = True
    return estat

def has_guanyat_diag1 (mat):
    """
    
    Description
    ----------
    Reads matrix and looks for 4 equivalent diagonal positions (left to right)
    
    Parameters
    ----------
    mat : matrix (list).

    Returns
    -------
    estat : True (4 equivalent positions)
            False (rest)

    """
    
    estat = False
    
    for i in range (0,3):
        for j in range (0,4):
            if mat [i][j] == mat [i+1][j+1] and mat [i][j] != " _ ":
                if mat [i][j] == mat [i+2][j+2] and mat [i][j] != " _ ":
                    if mat [i][j] == mat [i+3][j+3] and mat [i][j] != " _ ":
                        estat = True
    
    return estat


def has_guanyat_diag2 (mat):
    """
    
    Description
    ----------
    Reads matrix and looks for 4 equivalent diagonal positions (right to left)
    
    Parameters
    ----------
    mat : matrix (list).

    Returns
    -------
    estat : True (4 equivalent positions)
            False (rest)

    """
    
    estat = False
    
    for j in range (3,7):
        for i in range (0,3):
            if mat [i][j] == mat [i+1][j-1] and mat [i][j] != " _ ":
                if mat [i][j] == mat [i+2][j-2] and mat [i][j] != " _ ":
                    if mat [i][j] == mat [i+3][j-3] and mat [i][j] != " _ ":
                        estat = True
        return estat


def main ():
    """
    
    Description
    ----------
    Game methodology, shows matrix, lets player 1 choose column, shows modifyed 
    matrix, checks for 4 equivalences, lets computer choose column and checks again.
    
    The game is played with columns from 1 to 7

    Returns
    -------
    None.

    """
    
    j2 = matxbuida()
    taulell(j2)
    
    estat = False
    
    while estat == False:
        
        j1 = jugador1(j2)
      
        hgf1 = has_guanyat_files(j1)
        hgc1 = has_guanyat_cols(j1)
        hgd11 = has_guanyat_diag1(j1)
        hgd21 = has_guanyat_diag2(j1)
        
        j1 = taulell(j1)
      
        if (hgf1 or hgc1 or hgd11 or hgd21) == True:
            print ("")
            print ("Ben jugat 🎉")
            break

        print ("")
        j2 = jugador2(j1)
        
        hgf2 = has_guanyat_files(j2)
        hgc2 = has_guanyat_cols(j2)
        hgd12 = has_guanyat_diag1(j2)
        hgd22 = has_guanyat_diag2(j2)
        
        j2 = taulell(j2)
        
        if (hgf2 or hgc2 or hgd12 or hgd22) == True:
            print ("")
            print ("Mala sort 😢")
            break


if __name__=="__main__":
    main()
```
